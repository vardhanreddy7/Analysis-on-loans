# Data Cleaning
Now the next step is to clean the data. Data cleaning is used to get rid of unwanted rows, columns, missing data, etc.


```python
df.head(5)
```

```python
#let us now drop the coulmns that has all null values 
df = df.dropna(axis = 1, how = 'all')
```

```python
df.shape
```

```python
#finding the sum of null values in each column
df.isnull().sum()
```

- last_pymnt_d --> Last month payment was received
- next_pymnt_d --> Next scheduled payment date
- tax_liens --> Number of tax liens

There are almost 38577 loans that are closed

```python
# Let's now convert this sum into percentage for each columns as it would be easy to clean and sort the data that way

round(df.isnull().sum()/len(df)*100,2)
```

```python
# Drop columns with more than 60% null values
df=df.loc[:,round(df.isnull().sum()/len(df)*100,2)<60]
df.shape
```

```python
df.isnull().sum()
```

```python
data_dic = pd.read_excel("desktop/data/Data_Dictionary.xlsx")
```

```python
pd.set_option('display.max_colwidth', -1)
pd.set_option('display.max_columns', None)
data_dic[data_dic['LoanStatNew'].isin(list(df))]
```

```python
drop_columns=['id','member_id', 'emp_title', 'url', 'desc','title', 'earliest_cr_line',
              'last_pymnt_d','last_credit_pull_d','collection_recovery_fee', 'inq_last_6mths']
df=df.drop(drop_columns,axis=1)
```

```python
# Converting revol_util and int_rate into a numeric type
df.revol_util=df.revol_util.str.rstrip('%').astype('float')
df.int_rate=df.int_rate.str.rstrip('%').astype('float')

# Converting zip code to numbric by removing the X coded for braveity
df['zip_code'] = df['zip_code'].apply(lambda s: int(s[:-2]))
df["pub_rec_bankruptcies"].fillna(0.0, inplace=True)
df["revol_util"].fillna(0.0, inplace=True)
df.dtypes
```
